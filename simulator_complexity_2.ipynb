```python
import numpy as np
from numpy.random import default_rng
from decoder import *
```

```python
TRANSP_LEN = 10000
N_GEN = 100

rng = default_rng(0)
genome_size = 1000000000
```

```python
def modify_subgenome(subgenome, gen_size):
    new_transp = rng.integers(low=0, high=gen_size - TRANSP_LEN + 1, size=1)
    new_subgenome = subgenome
    for i, transp in enumerate(subgenome):
        if new_transp in range(transp, transp + TRANSP_LEN):
            new_subgenome = np.delete(new_subgenome, i)
        if transp > new_transp:
            new_subgenome[i] = transp + TRANSP_LEN
    gen_size += TRANSP_LEN
    new_subgenome = np.sort(np.concatenate([subgenome, new_transp]))
    return new_subgenome, gen_size
```

```python
def time_lapse(n_gens, subgenome, genome_size):
    for _ in range(n_gens):
        subgenome, genome_size = modify_subgenome(subgenome, genome_size)
        
    return subgenome, genome_size
```

```python
subgenome1 = rng.integers(low=0, high=genome_size - TRANSP_LEN + 1, size=5)
subgenome2 = subgenome1
subgenome3 = subgenome1

new_subgenome1 = time_lapse(10, subgenome1, genome_size)[0]
new_subgenome2 = time_lapse(10, subgenome2, genome_size)[0]
new_subgenome3 = time_lapse(10, subgenome3, genome_size)[0]
```
