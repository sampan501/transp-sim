```python
import numpy as np
```

```python
#Create subgenome0
genome_size0=1000000
TE_size=10000
TE_n =10

# Nucleotides are represented with numbers: 0=A, 1=C, 2=G, 3=T
subgenome0 = list(np.random.randint(4, size=genome_size0-TE_n*TE_size)) 
TE_seq = list(np.random.randint(4, size=10000))

TE_pos0=[]
for _ in range(10): # insert 10 transposons into random positions
    pos = np.random.randint(len(subgenome0))
    TE_pos0.append(pos)
    for i in range(len(TE_seq)):
        subgenome0.insert(i+pos,TE_seq[i]) 
        
#Note: subgenome0 is the "original/ancestral" genome
```

```python
genomes=[[] for _ in range(4)] #create an empty list of lists 
# genome[0] = Giant Sequoia
# genome[1] = Redwood subgenome1
# genome[2] = Redwood subgenome2
# genome[3] = Redwood subgenome3

#subgenome0 diverges into sequoia and duplicates in redwood (WGD#1)
for i in range(3):
    subgenome=list(subgenome0) #copying subgenome0
    TE_jump = np.random.poisson(5)
    for _ in range(TE_jump):
        pos = np.random.randint(len(subgenome))
        for j in range(len(TE_seq)):
            subgenome.insert(j+pos,TE_seq[j])   
    genomes[i]=(list(subgenome))
```

```python
#After WGD#1, Redwood is 4n. 
#Next, we assume that one of the Redwood subgenomes duplicates again (WGD#2)
#For this simulation, we will assume that genomes[1] duplicates to produce genomes[3]

genomes[3]=list(genomes[1])

for i in range(len(genomes)):
    subgenome=list(genomes[i])
    TE_jump = np.random.poisson(5)
    for _ in range(TE_jump):
        pos = np.random.randint(len(subgenome))
        for j in range(len(TE_seq)):
            subgenome.insert(j+pos,TE_seq[j])   
    genomes[i]=(list(subgenome))
```

```python
for i in range(len(genomes)):
    print("The length of genome is %i" %i,len(genomes[i]))
```

```python
d={0:"A",1:"C",2:"G",3:"T"}

for i  in range(len(genomes)):
    var = map(lambda x: d[x],genomes[i])
    seq="".join(list(var))
    file_name = "subgenome_" + str(i) + ".fa"
    fa = open(file_name,'w')
    fa.write(">" + file_name + '\n' + seq)
    fa.close()  
    
#Write transposon sequence to file:
var = map(lambda x: d[x],TE_seq)
seq="".join(list(var))
fa = open('transposon.fa','w')
fa.write(">" +"transposon"+'\n' + seq)
fa.close()
```
