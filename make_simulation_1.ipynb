## Simulation 1 

- same genome replicated 3 times

```python
# decide on size of genome and transposon
length_genome = 100
length_transposon = 10
# create subgenome 1
sub_genome_label = 1
sub_g1 = np.ones(length_genome)
# insert transposon
random.seed(0)
start_t = int(random.random()*length_genome-length_transposon)
end_t = start_t + length_transposon
sub_g1[start_t] = 50
sub_g1[end_t] = 50
# replication event 1
sub_g11 = np.concatenate((sub_g1,sub_g1), axis=0)
# replication event 2
sim1_genome = np.concatenate((sub_g1,sub_g11),axis=0)
len(sim1_genome)
```
